```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
for col in categorical_columns:
    df[col] = label_encoder.fit_transform(df[col].astype(str))  # .astype(str) ensures no error with NaNs
print(df.head())
```

```python
for i in df.select_dtypes(include="number").columns:
    sns.boxplot(data=df,x=i,color="pink")
    plt.show()
```

```python
def wisker(col):
    q1,q3=np.percentile(col,[25,75])
    iqr=q3-q1
    lw=q1-1.5*iqr
    uw=q3+1.5*iqr
    return lw,uw
for i in df.select_dtypes(include="number").columns:
    lw,uw=wisker(df[i])
    df[i]=np.where(df[i]<lw,lw,df[i])
    df[i]=np.where(df[i]>uw,uw,df[i])
for i in df.select_dtypes(include="number").columns:
    sns.boxplot(data=df,x=i,color="turquoise")
    plt.show()
```

```python
df1 = df.dropna(axis=1)

# Show result
print("Remaining columns:")
print(df1.columns)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load the dataset
df2 = df1.copy()

# Select only numerical columns for PCA
numerical_columns = df2.select_dtypes(include=[np.number]).columns
X = df2[numerical_columns]

# Standardize the data (PCA is sensitive to scale)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Choose number of components (e.g., 95% variance)
explained_variance = np.cumsum(pca.explained_variance_ratio_)
num_components = np.argmax(explained_variance >= 0.95) + 1  # Find first component where variance >= 95%

print(f"Optimal number of components: {num_components}")

# Apply PCA with the chosen number of components
pca_final = PCA(n_components=num_components)
X_reduced = pca_final.fit_transform(X_scaled)

# Convert back to DataFrame with original column names appended with "_PCA"
df_pca = pd.DataFrame(X_reduced, columns=[f"PC{i+1}" for i in range(num_components)])
print("Transformed Dataset:")
print(df_pca.head())
```

```python

import pandas as pd
import numpy as np
import xgboost as xgb
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import OneHotEncoder  # For categorical encoding
import shap


# 3. Prepare Data for XGBoost:
X = df1.drop(columns=["Year","Not Suicide(0)/Suicide(1)"])  # Features (exclude target)
# Include 'Social Isolation' and "Year" column in the features
y = df1["Not Suicide(0)/Suicide(1)"]  # Target variable

# 4. Split Data:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

# 5. Train XGBoost Model:
model = xgb.XGBClassifier(
    objective='binary:logistic',
    max_depth=4,
    learning_rate=0.1,
    n_estimators=150,
    use_label_encoder=False,
    eval_metric='logloss'
)
model.fit(X_train, y_train)

# 6. Predict and Evaluate:
predictions = model.predict(X_test)
accuracy = accuracy_score(y_test, predictions)
print(f'Accuracy of the model is: {accuracy*100:.2f}%')

# 7. Analyze Feature Importance:

# a. Built-in Feature Importance:
print("XGBoost Feature Importance:")
print(model.feature_importances_)
xgb.plot_importance(model, importance_type='gain')  # Use 'gain' for better interpretation
plt.show()

# b. SHAP Values:
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X_test)

# Summary plot
shap.summary_plot(shap_values, X_test, feature_names=X_test.columns)

# Dependence plot for 'Social Isolation' (or its encoded features)
shap.dependence_plot('Social Iscolation', shap_values, X_test, feature_names=X_test.columns)
# If one-hot encoded, use the encoded feature name like 'Social Isolation_Yes'

# Force plot for a specific instance
shap.initjs()
shap.force_plot(explainer.expected_value, shap_values[0,:], X_test.iloc[0,:], feature_names=X_test.columns)
```
